```python
# Tabela de frequências absolutas

tab1 = pd.crosstab(dados['is_medicamento'], columns='count')
tab1
```

```python
tab1.plot.bar(color='#044152')  # Mantém as barras verticais e aplica a cor
plt.legend(title='count')
plt.xticks(rotation=0)  # Mantém os rótulos do eixo X na horizontal
plt.show()
```

```python
# Tabela de frequências absolutas

cores = ['#044152', '#10f35b', '#e4f6ea', '#2ef156', '#139f52']

tab = pd.crosstab(index=dados['curva'], columns='count')
tab

tab.plot.bar(color=cores)
plt.legend(title='Curvas')

plt.show()
```

```python
# Definir a paleta personalizada
linqi_palette = ["#044152", "#10f35b", "#e4f6ea", "#2ef156", "#139f52", 
                 "#6dee9b", "#4ee37f", "#84fcb4", "#2ce944", "#506c6c"]

# Aplicar a paleta aos gráficos Seaborn
sns.set_palette(sns.color_palette(linqi_palette))
```

```python
var_quali = dados[['loja_id','is_medicamento', 'curva']]

# Gráficos para variáveis qualitativas
for col in var_quali.columns:
    plt.figure()
    sns.countplot(data=dados, x=col, palette=linqi_palette)
    plt.xticks(rotation=45)
    plt.title(f'Distribuição de {col}')
    plt.show()
```

 <font color = #044152> **Gráfico de dispersão**

```python
sns.set_palette('colorblind')
sns.relplot(x='estoque', y='venda', hue='is_medicamento', col='curva',  data=dados)
```
